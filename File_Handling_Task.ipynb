```python
import datetime
error = '\nWrong Input\nTry Again\n'
now = str(datetime.datetime.now().strftime("%D %H:%M:%S"))

#Creating Data
def create_loki_diet():
  global f1,f2,f3,f4
  f1 = open('loki_diet','a')
def create_loki_excercise():
  global f1,f2,f3,f4
  f2 = open('loki_excercise','a')
def create_clint_diet():
  global f1,f2,f3,f4
  f3 = open('clint_diet','a')    
def create_clint_excercise():
  global f1,f2,f3,f4
  f4 = open('clint_excercise','a')
  
#Writing Data
def write_loki_diet(message):
  create_loki_diet()
  f1.write("\n"+now+"\t"+message)
  print('\nExcercise added!')
def write_loki_excercise(message):
  create_loki_excercise()
  f2.write("\n"+now+"\t"+message)
  print('\nExcercise added!')
def write_clint_diet(message):
  create_clint_diet()
  f3.write("\n"+now+"\t"+message)
  print('\nDiet added!')
def write_clint_excercise(message):
  create_clint_excercise()
  f4.write("\n"+now+"\t"+message)
  print('\nExcercise added!')

#Reading Data 
def read_loki_diet():
  f1 = open('loki_diet','r')
  print("\n"+f1.read())
def read_loki_excercise():
  f2 = open('loki_excercise','r')
  print("\n"+f2.read())
def read_clint_diet():
  f3 = open('clint_diet','r')
  print("\n"+f3.read())
def read_clint_excercise():
  f4 = open('clint_excercise','r')
  print("\n"+f4.read())
  
breaker = False 
while True: 
  while True:
    person = input("\n1)Loki\n2)Clint\nQ - to Quit: ")
    if person == '1':
      while True:
        lifestyle = input("\nLoki:\n1)Diet\n2)Excercise\n3)Previous Menu: ")
        if lifestyle == '3':
          break
        elif lifestyle == '1':
          create_loki_diet()
          while True:
            action = input("\nLoki's Diet:\n1)Add\n2)Read\n3)Previous Menu\n")
            if action == '3':
              break
            elif action == '1':
              while True:
                x = input("Enter an Item to Loki's Diet: ")
                write_loki_diet(x)
                more = input("\nWould you like to add more items to Loki's diet?\n1)Yes 2)No: ")
                if(more=='1'):
                  continue
                elif(more == '2'):
                  f1.close()
                  break
            elif action == '2':
              read_loki_diet()
            else:
              print(error)
              continue
        elif lifestyle == '2':
          create_loki_excercise()
          while True:
            action = input("\nLokis Excercise:\n1)Add\n2)Read\n3)Previous Menu\n")
            if action == '3':
              break
            elif action == '1':
              while True:
                x = input("Enter an Item to Loki's Excercise: ")
                write_loki_excercise(x)
                more = input("\nWould you like to add more items to loki's Excercise?\n1)Yes 2)No: ")
                if(more=='1'):
                  continue
                elif(more == '2'):
                  f2.close()
                  break
            elif action == '2':
              read_loki_excercise()
            else:
              print(error)
              continue
        else:
          print(error)                
    elif person == '2':
      while True:
        lifestyle = input("\nClint:\n1)Diet\n2)Excercise\n3)Previous Menu: ")
        if lifestyle == '3':
          break
        elif lifestyle == '1':
          create_clint_diet()
          while True:
            action = input("\nClint's Diet:\n1)Add\n2)Read\n3)Previous Menu\n")
            if action == '3':
              break
            elif action == '1':
              while True:
                x = input("Enter an Item to Clint's Diet: ")
                write_clint_diet(x)
                more = input("\nWould you like to add more items to Clint's diet?\n1)Yes 2)No: ")
                if(more=='1'):
                  continue
                elif(more == '2'):
                  f3.close()
                  break
            elif action == '2':
              read_clint_diet()
            else:
              print(error)
              continue
        elif lifestyle == '2':
          create_clint_excercise()
          while True:
            action = input("\nClint's Excercise:\n1)Add\n2)Read\n3)Previous Menu\n")
            if action == '3':
              break
            elif action == '1':
              while True:
                x = input("Enter an Item to Clint's Excercise: ")
                write_clint_excercise(x)
                more = input("\nWould you like to add more items to Clint's Excercise?\n1)Yes 2)No: ")
                if(more=='1'):
                  continue
                elif(more == '2'):
                  f4.close()
                  break
            elif action == '2':
              read_clint_excercise()
            else:
              print(error)
              continue
        else:
          print(error)    
    elif person =='Q' or person =='q':
      breaker = True
      break
    else:
      print(error)    
  if breaker == True:
    break
```
